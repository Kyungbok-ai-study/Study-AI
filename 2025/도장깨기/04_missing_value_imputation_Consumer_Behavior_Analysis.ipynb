```python
file_path = '/content/drive/MyDrive/data/processed_df.pkl'
import pandas as pd
df = pd.read_pickle(file_path)

df
```

```python
# prompt: df의 모든 컬럼의 데이터형태, 고유값, 결측치 개수와 비율, 각 컬럼의 10개의 값을 보여주라. 데이터프레임 형태로 간단하게

import pandas as pd

def df_info(df):
    info_dict = {
        'column': [],
        'dtype': [],
        'unique_values': [],
        'missing_count': [],
        'missing_ratio': [],
        'first_10_values': []
    }

    for col in df.columns:
      info_dict['column'].append(col)
      info_dict['dtype'].append(df[col].dtype)
      info_dict['unique_values'].append(df[col].nunique())
      missing_count = df[col].isnull().sum()
      info_dict['missing_count'].append(missing_count)
      info_dict['missing_ratio'].append(missing_count / len(df) * 100)
      info_dict['first_10_values'].append(df[col].unique()[:10].tolist())

    return pd.DataFrame(info_dict)

df_info(df)
```

```python
# prompt: 자녀수의 결측치와  무응답을 자녀수의 최빈값으로 대체하는 코드

# Fill missing values and "무응답" in '자녀수' with the mode
df['자녀수'] = df['자녀수'].fillna(df['자녀수'].mode()[0])
df['자녀수'] = df['자녀수'].replace('무응답', df['자녀수'].mode()[0])
```

```python
df_info(df)
```

```python
# prompt: 결측치가 있는 모든 컬럼은 최빈값으로 대체하라

for col in df.columns:
  if df[col].isnull().any():
    mode_val = df[col].mode()[0]
    df[col] = df[col].fillna(mode_val)

df_info(df)
```

```python
# prompt: object 형태의 컬럼을 category로 변환

# Check object columns
object_cols = df.select_dtypes(include='object').columns

# Convert object columns to category type
for col in object_cols:
    df[col] = df[col].astype('category')

df_info(df)
```

```python
df['PPL구입여부'].map({'있다': 1, '없다': 0}).value_counts()
```

```python
# prompt: PPL구입여부의  있다, 없다 값을 1과 0 정수로 변환하라

# Convert 'PPL구입여부' values to numerical representation
df['PPL구입여부'] = df['PPL구입여부'].map({'있다': 1, '없다': 0})

df_info(df)
```

```python
# prompt: 남자와 여자 구분에 따른 구매여부를 2줄 코드로

import pandas as pd

# Assuming 'df' is your DataFrame and it has columns '성별' and 'PPL구입여부'
purchase_by_gender = df.groupby('성별')['PPL구입여부'].value_counts(normalize=True) * 100
purchase_by_gender
```

```python
# prompt: year에 따라 성별의 구매여부가 달라지는가?

# Group by 'year' and '성별' and calculate the purchase rate
purchase_by_year_gender = df.groupby(['year', '성별'])['PPL구입여부'].value_counts(normalize=True) * 100

# Display the results
purchase_by_year_gender
```
